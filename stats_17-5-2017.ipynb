###### Abychom nic nepokazily, vytvorime novou tabulku "soubor1"

```python
soubor1 = soubor.loc[:,['Období', 'ATC7', 'Počet denních definovaných dávek/balení celkem']]
```

```python
soubor1 = soubor1.rename(index=str, columns={'Počet denních definovaných dávek/balení celkem': 'DDD_total'})
soubor1.head()
```

```python
''' 
Chceme pouzit GROUPBY:
leky = soubor1.groupby(['Období','ATC7']).agg({'Počet denních definovaných dávek/balení celkem':'sum'})
leky.head(3)

Ale bohuzel to leky.columns vrati jen jeden sloupec: 
Index(['Počet denních definovaných dávek/balení celkem'], dtype='object')

Tak pridame dva dalsi
'''
```

```python
leky1 = soubor1.groupby(['Období','ATC7']).agg({'Období':'first','ATC7':'all', 'DDD_total':'sum'})#.unstack()
leky1.tail()
```

```python
leky1.to_csv('ATC_sum.csv', encoding = 'utf-8')
```

```python
leky1['Období'] = pd.to_datetime(leky1['Období'])
```

```python
leky1.tail(3)
```

```python
leky1['Období'] = pd.to_datetime(leky1['Období'])
```

```python
leky1['year'] = leky1['Období'].dt.year.astype(np.int)
```

```python
leky1['month'] =leky1['Období'].dt.month.astype(np.int)
```

```python
leky1['quater'] = leky1['Období'].dt.quarter.astype(np.int)
#DatetimeIndex.quarter
```

```python
leky1['trend'] = leky1['year']-2011
```

```python
leky1.tail(3)
```

### Pouzijeme get_dummies na mesici a kvartaly 

```python
Q1 = pd.get_dummies(leky1['quater'], prefix='quater')
Q1.head()
```

```python
M1 = pd.get_dummies(leky1['month'], prefix='month')
M1.head()
```

### Ted to spojime do jednoho dataframu "model_df"

```python
model_df = pd.concat([leky1, M1, Q1], axis=1, join_axes=[leky1.index])
model_df.head(3)
```

```python
model_df.info()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

### Vybereme jen jednu molekulu 'M01AE01' a vytvorime dataframe "model_df_M01AE01":

```python
model_df_M01AE01 = model_df[model_df['ATC7'] == 'M01AE01']
model_df_M01AE01.head()
```

### Napred "quarters":

```python
result_M1 = smf.ols('DDD_total ~  quater_1 + quater_2 + quater_3', data=model_df_M01AE01).fit()
print(result_M1.summary())
```

### Je to horsi, ze "nic moc"... Tak zkusime po mesicich:

```python
result_M01 = smf.ols('DDD_total ~  month_1 + month_2 + month_3 + month_4 + month_5 + month_6 + month_7 + month_8 + month_9 + month_10 + month_11', data=model_df_M01AE01).fit()
print(result_M01.summary())
```

### Zkusime druhou nejpouzivanejsi molekulu: N02BE01

```python
model_df_N02BE01 = model_df[model_df['ATC7'] == 'N02BE01']
model_df_N02BE01.head(3)
```

```python
resultsN = smf.ols('DDD_total ~ trend + quater_1 + quater_2 + quater_3', data= model_df_N02BE01).fit()
print(resultsN.summary())
```

```python
resultsN1 = smf.ols('DDD_total ~ trend +  month_1 + month_2 + month_3 + month_4 + month_5 + month_6 + month_7 + month_8 + month_9 + month_10 + month_11', data= model_df_N02BE01).fit()
print(resultsN1.summary())
```
